```python
import requests
import pandas as pd 
from requests.utils import quote
from pandas import ExcelWriter


df = pd.read_excel("address.xlsx",)
address = df['address']
address_list = []
for line in address:
    lat_lng = {}
    location = quote(line.strip())
#     enter api key to run
    key=""
    url = "https://maps.googleapis.com/maps/api/geocode/json?address="+location+"&key="+key
    r = requests.get(url)
    c = r.json()
    
    try:
        lat_lng["address"] = line
    except:
        lat_lng["address"] = " "
        
    try:
        lat_lng["lat"] = c["results"][0]["geometry"]["location"]["lat"]
    except:
        lat_lng["lat"] = " "
        
    try:
        lat_lng["lng"] = c["results"][0]["geometry"]["location"]["lng"]
    except:
        lat_lng["lng"] = " "
    
    address_list.append(lat_lng)

# print(address_list)
mydf = pd.DataFrame(address_list)
writer = ExcelWriter('locations122718.xlsx')
mydf.to_excel(writer,'Sheet1')
writer.save()
print("Done!")
mydf

    



```
